Data cleaning and preprocessing

```python
adidas_sales.info()
```

```python
adidas_sales = adidas_sales.drop_duplicates()
```

```python
adidas_sales["Invoice Date"] = pd.to_datetime(adidas_sales["Invoice Date"])
```

```python
coloms = ["Price per Unit","Units Sold","Total Sales","Operating Profit","Operating Margin"]
adidas_sales[coloms] = adidas_sales[coloms].astype(float)
```

```python
adidas_sales.info()
```

```python
adidas_sales["Retailer ID"] = adidas_sales["Retailer ID"].astype(int)
```

Exploratoty Data Analysis(EDA)

Sales By Reatailers

```python
adidas_sales["Retailer"].unique()
```

```python
revenue_by_Retailer = adidas_sales.groupby("Retailer")['Total Sales'].sum()
```

```python
fig, ax=plt.subplots()
ax.bar(revenue_by_Retailer.index , revenue_by_Retailer, color='#40D19A')
ax.set_xlabel("Retailers")
ax.set_ylabel("Sales")
ax.set_title("Total Sales By Retailers")
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
for i, v in enumerate(revenue_by_Retailer):
        ax.text(i, v, str(v), ha='center', va='bottom')
```

```python
fig, ax=plt.subplots()
ax.pie(revenue_by_Retailer, startangle=90, labels=revenue_by_Retailer.index, autopct="%.0f%%")
ax.set_title("Total Sales By Retailers")
plt.show()
```

Sales By Products

```python
adidas_sales["Product"].unique()
```

```python
revenue_by_Product = adidas_sales.groupby("Product")['Total Sales'].sum()
```

```python
fig, ax=plt.subplots()
ax.bar(revenue_by_Product.index , revenue_by_Product, color='#40D19A')
ax.set_xlabel("Product")
ax.set_ylabel("Sales")
ax.set_title("Total Sales By Product")
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
for i, v in enumerate(revenue_by_Product):
        ax.text(i, v, str(v), ha='center', va='bottom')
plt.xticks(rotation=90)
plt.show()
```

```python
fig, ax=plt.subplots()
ax.pie(revenue_by_Product, startangle=90, labels=revenue_by_Product.index, autopct="%.0f%%")
ax.set_title("Total Sales By Product")
plt.show()
```

Profit By Each Retailers and Each Product 

```python
profit_by_Retailer = adidas_sales.groupby("Retailer")['Operating Profit'].sum()
profit_by_Product = adidas_sales.groupby("Product")['Operating Profit'].sum()
```

```python
fig, ax=plt.subplots(2,2,figsize=(15,10))



ax[0][0].bar(profit_by_Retailer.index , profit_by_Retailer, color='#40D19A')
ax[0][0].set_xlabel("Retailers")
ax[0][0].set_ylabel("Sales")
ax[0][0].set_title("Total Profit By Each Retailers")
ax[0][0].spines['right'].set_visible(False)
ax[0][0].spines['top'].set_visible(False)
for i, v in enumerate(profit_by_Retailer):
        ax[0][0].text(i, v, str(v), ha='center', va='bottom')

ax[1][0].pie(profit_by_Retailer, startangle=90, labels=profit_by_Retailer.index, autopct="%.0f%%")
ax[1][0].set_title("Total Profit By Each Retailers")
        

    
ax[0][1].bar(profit_by_Product.index , profit_by_Product, color='#40D19A')
ax[0][1].set_xlabel("Products")
ax[0][1].set_ylabel("Sales")
ax[0][1].set_title("Total Profit By Each Products")
ax[0][1].spines['right'].set_visible(False)
ax[0][1].spines['top'].set_visible(False)
for i, v in enumerate(profit_by_Product):
        ax[0][1].text(i, v, str(v), ha='center', va='bottom')

ax[1][1].pie(profit_by_Product, startangle=90, labels=profit_by_Product.index, autopct="%.0f%%")
ax[1][1].set_title("Total Profit By Each Products")

plt.show()
```
